```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data={'Math':20,'Biology':15,'Commerce':25,'Arts':35}
subjects=list(data.keys())
scount=list(data.values())
```

```python
# size of figure/window:
fig=plt.figure(figsize=(10,5))   
```

```python
#bar graph:
plt.bar(subjects,scount,color='red',width=0.4)    #(x-axis,y-axis,....)

plt.xlabel('Subjects')  #x-axis label
plt.ylabel('Frequency of Students')  #y-axis label
plt.title('No. of students in each subject stream')   #title for plot

plt.show    #to print
```

```python
#lien graph:
plt.plot(subjects,scount,color='red')    #(x-axis,y-axis,....)

plt.xlabel('Subjects') 
plt.ylabel('Frequency of Students')  
plt.title('No. of students in each subject stream') 

plt.show  
```

```python
#scattter graph:
plt.scatter(subjects,scount,color='blue') 

plt.xlabel('Subjects') 
plt.ylabel('Frequency of Students')
plt.title('No. of students in each subject stream') 

plt.show 
```

```python
#pie graph:
plt.pie(scount,labels=subjects)    #(y-axis,labels=x-axis)

plt.title('No. of students in each subject stream')   

plt.show  
```

```python
temps={'Jan':10,'Feb':15,'Mar':22,'Apr':30}
humds={'Jan':76,'Feb':78,'Mar':86,'Apr':88}
months=list(temps.keys())
temp=list(temps.values())
humd=list(humds.values())
fig=plt.figure(figsize=(10,5))   

plt.plot(months,temp,color='red')
plt.plot(months,humd,color='blue')       #(x-axis,y-axis,....)

plt.xlabel('Months') 
plt.ylabel('Temp and Humd')  
plt.title('Avg. temperature and humidity for each month') 

plt.show  
```

```python
temps={'Jan':10,'Feb':15,'Mar':22,'Apr':30}
humds={'Jan':76,'Feb':78,'Mar':86,'Apr':88}
months=list(temps.keys())
temp=list(temps.values())
humd=list(humds.values())
fig=plt.figure(figsize=(10,5))   

x_axis=np.arange(len(months))

plt.bar(x_axis-0.1,temp,label='Temp.',width=0.2)  #(shifitng axis, x-axis, y-axis)
plt.bar(x_axis+0.1,humd,label='Humidity',width=0.2)  

plt.xticks(x_axis,months)    # for x axis labels

plt.grid(True,'major','y')

plt.xlabel('Months') 
plt.ylabel('Temp and Humd')  
plt.title('Avg. temperature and humidity for each month') 

plt.show  
```
